```python
import sys 
from PyQt5.QtWidgets import QMessageBox

from PyQt5 import uic,QtCore
from PyQt5.QtWidgets import *
from PyQt5 import *
class Calculator(QDialog):
    def __init__(self):
        super(Calculator,self).__init__()
        uic.loadUi("calculator.ui",self)
        self.but0.clicked.connect(self.z) 
        self.but1.clicked.connect(self.o) 
        self.but2.clicked.connect(self.t) 
        self.but3.clicked.connect(self.th) 
        self.but4.clicked.connect(self.f) 
        self.but5.clicked.connect(self.fi) 
        self.but6.clicked.connect(self.s) 
        self.but7.clicked.connect(self.se) 
        self.but8.clicked.connect(self.e)
        self.but9.clicked.connect(self.n)
        self.butminus.clicked.connect(self.mi)
        self.butplus.clicked.connect(self.pl)
        self.butmulti.clicked.connect(self.mu)
        self.butdiv.clicked.connect(self.di)
        self.butequal.clicked.connect(self.eq)
        self.butclear.clicked.connect(self.cl)
        
    def z(self):
         text=self.label.text()
         self.label.setText(text+"0")
         self.msgshow("number","You clicked 0")
         
         
    def o(self):
         text=self.label.text()
         self.label.setText(text+"1")
         self.msgshow("number","You clicked 1")
    
    def t(self):
         text=self.label.text()
         self.label.setText(text+"2")
         self.msgshow("number","You clicked 2")
    def th(self):
         text=self.label.text()
         self.label.setText(text+"3")
         self.msgshow("number","You clicked 3")
    def f(self):
         text=self.label.text()
         self.label.setText(text+"4")
         self.msgshow("number","You clicked 4")
    def fi(self):
         text=self.label.text()
         self.label.setText(text+"5")
         self.msgshow("number","You clicked 5")
    def s(self):
        text=self.label.text()
        self.label.setText(text+"6")
        self.msgshow("number","You clicked 6")
    def se(self):
         text=self.label.text()
         self.label.setText(text+"7")
         self.msgshow("number","You clicked 7")
         
    def e(self):
         text=self.label.text()
         self.label.setText(text+"8")
         self.msgshow("number","You clicked 8")
         
    def n(self):
        text=self.label.text()
        self.label.setText(text+"9")
        self.msgshow("number","You clicked 9")
    def mi(self):
         text=self.label.text()
         self.label.setText(text+"-")
         self.msgshow("sign","You clicked -")
    def pl(self):
         text=self.label.text()
         self.label.setText(text+"+")
         self.msgshow("sign","You clicked +")
    def mu(self):
         text=self.label.text()
         self.label.setText(text+"*")
         self.msgshow("sign","You clicked *")
    def di(self):
         text=self.label.text()
         self.label.setText(text+"/")
         self.msgshow("sign","You clicked /")
    def cl(self):
         self.label.setText(" ")
         self.msgshow("sign","You clicked clear") 
    def eq(self):
        equation=self.label.text()
        self.msgshow("sign","You clicked equal")
        try:
            ans=eval(equation)
            self.label.setText(str(ans))
        except:
            self.label.setText("Error")
            
               
    def msgshow(Self,title,body):
         mesg=QMessageBox()
         mesg.setWindowTitle(title)
         mesg.setText(body)
         mesg.exec()
   
    
    
    
app=QApplication([])
win=Calculator()
win.show()
sys.exit(app.exec())
```
